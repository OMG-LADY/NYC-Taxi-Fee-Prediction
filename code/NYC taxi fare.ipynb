```python
def distance(lat1,lon1,lat2,lon2):
    from math import cos, asin, sqrt
    p = 0.017453292519943295  #pi/180
    a = 0.5-((lat2-lat1)*p).apply(cos)/2+(lat1*p).apply(cos)*(lat2*p).apply(cos)*(1-((lon2-lon1)*p).apply(cos))/2
    return 12742*(a.apply(sqrt)).apply(asin) #km
```

```python
train['dist_line'] = distance(train['pickup_latitude'],train['pickup_longitude'],train['dropoff_latitude'],train['dropoff_longitude'])
```

After making a scatter plot of dist_line and fare_amount, there're lots of outliers so we'll drop them before observing the correlation pattern.

```python
# drop outliers
# drop the rows that have fare_amount less than 0
train = train[train['fare_amount'] > 0]
# drop the rows that have 0 dist_line (which means car didn't move while some have fare_amount, some geocodes are 0)
train = train[train['dist_line'] != 0]
```

```python
# rows with fare_amount larger than 500 seems to be problematic data too so we will drop them
train[train['fare_amount'] > 500]
```

```python
# drop the rows that have fare_amount larger than 500
train = train[train['fare_amount'] <= 500]
```

```python
# drop the rows that have 0 in either of the four geocodes
train = train[(train['pickup_longitude'] != 0) & (train['pickup_latitude'] != 0) & (train['dropoff_longitude'] != 0) & (train['dropoff_latitude'] != 0)]
```

```python
# drop the rows that have dist_line > fare_amount since it only makes sense if fare_amount > dist_line
train = train[train['fare_amount'] > train['dist_line']]
```

```python
train[['fare_amount']].describe()
```

```python
train[['dist_line']].describe()
```

```python
train.plot(kind='scatter',x='dist_line',y='fare_amount',alpha=0.5)
plt.show()
```

```python
# check the rows that have dist_line < 10 while fare_amount > 200 (makes no sense)
train[(train['fare_amount'] > 200) & (train['dist_line'] < 10)]
```

```python
# drop these rows
train.drop(train[(train['fare_amount'] > 200) & (train['dist_line'] < 10)].index,inplace=True)
```

### Try Linear Regression

1. Use only dist_line to predict the fare_amount

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
reg = linear_model.LinearRegression()
```

```python
x = train[['dist_line']]
y = train[['fare_amount']]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
reg.fit(x_train,y_train)
y_pred = reg.predict(x_test)
reg.score(x_test,y_test) #R^2 score
```

```python
reg.coef_
```

```python
# use the whole train data to train the model then predict the fare_amount on test set
reg_all = linear_model.LinearRegression()
reg_all.fit(x,y)
```
